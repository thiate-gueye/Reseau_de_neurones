```python
def est_premier(a):
    for i in range(2,(a//2 +1) ):
        if a%i == 0:
            return False
    return True;
```

```python
def premier(nb):
    a = 2;
    listes = []
    while (a <= nb):
        if (est_premier(a)):
            listes.append(a)
        a = a + 1;
    return listes;
```

```python
def mes_premier(nb):
    listes = premier(nb)
    premiers = []
    for a in listes:
        if(nb % a != 0):
            premiers.append(a)
    return premiers;
```

```python
def p_et_q(n):
    listes = []
    pre = premier(n/2)
    a = 0 
    l = len(pre)
    while(a < l): 
        for i in range(a,l):
            if(i*a == n):
                listes.append(i)
                listes.append(a)
                return listes
        a = a + 1
    return listes
```

```python
def congrue(phi, e):
    cong = dict()
    i = 1
    while(i<=1000):
        if((phi*i +1) % e == 0 ):
            cong[i] = (phi*i +1) / e
        i = i + 1
    return cong
```

```python
pq = p_et_q(187)
```

```python
phi_de_n = (pq[0]-1)*(pq[1]-1)
```

```python
pre = mes_premier(phi_de_n)
```
