```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
```

```python
X,Y = make_blobs(n_samples=200,n_features=2,centers=2,random_state=0)
Y = Y.reshape(Y.shape[0],1)
print("dimension de x : ",X.shape)
print("dimension de y : ",Y.shape)

plt.scatter(X[:,0],X[:,1] ,c=Y)
plt.show()
```

```python
def initialisation(X):
    W = np.random.randn(X.shape[1],1)
    b = np.random.randn(1)
    return (W,b)
```

```python
W,b = initialisation(X) 
```

```python
def model(X,W,b):
    Z = X.dot(W) + b
    A = 1 / (1+ np.exp(-Z))
    return A
```

```python
A = model(X,W,b)
```

```python
def log_loss(A,Y):
    return 1 / len(Y) * np.sum(-Y * np.log(A) - (1-Y) * np.log(1-A))
```

```python
def gradiants(A,X,Y):
    dW = 1 / len(Y) * np.dot(X.T , A - Y)
    db = 1 / len(Y) * np.sum(A - Y)
    return dW, db
```

```python
dW, db = gradiants(A,X,Y)
```

```python
def update(dW, db, W, b, l_rate):
    W = W - l_rate * dW
    b = b - l_rate * db
    return (W,b) 
```

```python
A,B = update(dW, db, W, b, 0.01)
```

```python
def prediction(X,W,b):
    A = model(X,W,b)
    print(A)
    return A >= 0.5
```

```python
def artifical_neuron(X,Y,l_rate = 0.1, n_iter = 200):
    #initialisation W,b
    W,b = initialisation(X)
    loss = []
    for i in range(n_iter):
        A = model(X,W,b)
        loss.append(log_loss(A,Y))
        dW, db = gradiants(A,X,Y) 
        W,b = update(dW, db, W, b, l_rate)
    predict = prediction(X,W,b)
    print(accuracy_score(Y,predict))
    plt.plot(loss)
    return W,b
```

```python
W,b = artifical_neuron(X,Y)
```

```python
new_plant = np.array([2,1])
new_plant2 = np.array([2,3])
```

```python
x0 = np.linspace(-1,4,200)
x1 = (-W[0] * x0 -b) / W[1]
plt.scatter(X[:,0],X[:,1] ,c=Y)
plt.scatter(new_plant[0],new_plant[1] ,c='r')
plt.scatter(new_plant2[0],new_plant2[1] ,c='blue')
plt.plot(x0,x1,c="green",lw=2)
plt.show()
```

```python
print(prediction(new_plant,W,b))
print(prediction(new_plant2,W,b))
```
